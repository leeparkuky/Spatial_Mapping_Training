```python
class FIPSdata:
    def __init__(self):
        # read the pickle file
        #self.data = pickle.load(open(pickle_file_path, 'rb'))
        #self.remove_index()
        pass
        
    @classmethod
    def read_dict(cls, Dict):
        obj = cls()
        obj.data = Dict
        obj.data = {key:value for (key, value) in obj.data.items() if '_long' not in key}   
        obj.remove_index()
        return(obj)

        
    @classmethod
    def read_pickle(cls, filepath):
        obj = cls()
        obj.data = pickle.load(open(filepath, 'rb'))
        obj.data = {key:value for (key, value) in obj.data.items() if '_long' not in key}   
        obj.remove_index()
        return(obj)
        
        
    def remove_index(self):
        data = self.data
        for k,v in data.items():
            newdf = v.reset_index()
            if newdf.columns.str.contains('index').sum():
                #newdf = newdf.drop("index", axis = 1)   # axis = 1 means columns
                #newdf.drop("index", axis = 1, inplace = True)
                pass
            else:
                #self.data[k] = newdf
                self.data[k] = newdf
    
    # want ot find unique state FIPS in a data dictinoary > [33,?]
    @property
    def state(self):
        state=[]
        data = self.data
        #for each data.values(), if it has FIPS column, then use np.unique. 
        for k,v in data.items():
            if k not in ['facilities_and_providers', 'broadband_speeds']: # only two keys are not what we need
                self._state = v.FIPS.apply(lambda x:str(x)[0:2]).unique()
            
#             self._state = np.unique([int(i[:2]) for i in j.FIPS if j.FIPS and i[:2]]).tolist()

            
    def split_data(self):
        self.county = {key:value for (key, value) in self.data.items() if 'county' in key}
        self.tract = {key:value for (key, value) in self.data.items() if 'tract' in key}
        self.other = {key:value for (key, value) in self.data.items() if 'county' not in key and 'tract' not in key}
        
    
    @staticmethod
    def print_hello(message):
        print(f"hello user {message}")
```
