Base map: like the topology map bottom layer
Feature layer: like the "clear plastic" over the topology map with features marked

  * Area variable: indicates a whole geographic region (collection of point variables)
  * Point variable: indicates a single point (for example using latitude/longitude notation)
  * Map: the "bundle" of base and feature layers
